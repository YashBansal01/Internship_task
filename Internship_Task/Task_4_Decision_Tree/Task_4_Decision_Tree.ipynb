```python
# Importing libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

```

```python
#Importing Given Dataset
df=pd.read_csv("iris.csv")

```

```python
#Getting Information About The Dataset
df.info
```

```python
#Glancing The DataSet
df.head()
```

```python
#Checking for presence of NUll Values
df.isnull().sum()
```

```python
#Getting the input and output values from Dataset
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Y = df[['Species']]
```

```python
#Splitting the Dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size = 0.2 , random_state=2018)
```

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X_train, Y_train)

print('Decision Tree Classifer Created')
```

```python
dtree_pred = dtree.predict(X_test)
```

```python
#Evaluating the Model
from sklearn.metrics import accuracy_score, classification_report
print("Accuracy Score Decision Tree" , accuracy_score(Y_test , dtree_pred))
print("Classification Report Decision Tree" , classification_report(Y_test , dtree_pred))
```

```python
#Visualizing the Decision Tree Model
from sklearn import tree
fn =['SepalLength(Cm)','SepalWidth(Cm)','PetalLength(Cm)','PetalWidth(Cm)']
cn = ['sentosa','versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1 ,ncols = 1,figsize = (4,4), dpi = 300)
tree.plot_tree(dtree,feature_names = fn, class_names = cn, filled = True);
```
